# Writing and executing a simple Python function

We define a simple function called `greet` and test it with several arguments.

```python
def greet(name: str) -> str:

    """Return a friendly greeting for the given name."""

    return f'Hello, {name}!'



# Test the function

greet('World')
```

```python
# More usage examples

for name in ['Ana', 'Luis', 'Data Scientist']:

    print(greet(name))
```

# Reading and writing files from the notebook

We use standard Python to create a text file, write content, and read it back.

```python
import os



# Show the current working directory of the notebook

print('Current working directory:', os.getcwd())



# Define a relative file path

file_path = 'example_file.txt'



# Write content to the file

with open(file_path, 'w', encoding='utf-8') as f:

    f.write('This is an example file created from the notebook.\n')

    f.write('We can write multiple lines of text.\n')



# Read the file content

with open(file_path, 'r', encoding='utf-8') as f:

    content = f.read()



print('File content:')

print(content)
```

# Basic plotting with Matplotlib

We generate synthetic data (a sine wave) and draw it as a line plot.

```python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt



# Create example data: a sine wave

x = np.linspace(0, 2 * np.pi, 500)

y = np.sin(3 * x)



plt.figure(figsize=(8, 4))

plt.plot(x, y, label='sin(3x)')

plt.title('Example plot with Matplotlib')

plt.xlabel('x')

plt.ylabel('sin(3x)')

plt.legend()

plt.grid(True)

plt.show()
```

# Running unit tests from the notebook

We create functions to test and define tests using the `unittest` module. Then we show how to run the tests.

```python
# Functions under test

def add(a: int, b: int) -> int:

    return a + b



def is_even(n: int) -> bool:

    return n % 2 == 0
```

```python
# Create tests with unittest

import unittest



class TestFunctions(unittest.TestCase):

    def test_add(self):

        self.assertEqual(add(2, 3), 5)

        self.assertEqual(add(-1, 1), 0)



    def test_is_even(self):

        self.assertTrue(is_even(2))

        self.assertTrue(is_even(0))

        self.assertFalse(is_even(3))



# Run the tests from the notebook

suite = unittest.defaultTestLoader.loadTestsFromTestCase(TestFunctions)

runner = unittest.TextTestRunner(verbosity=2)

runner.run(suite)
```
